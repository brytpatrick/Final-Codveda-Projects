### Check Missing Values
This checks whether any information is missing from the dataset.

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\umani\Downloads\churn-bigml-80.csv")
dfprint(df.isnull().sum())
```

### Fill Missing Values
If any values are missing, numerical values are replaced with the median, while text values are replaced with the most common value.

```python
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode([0]))
```

### Remove duplicates
Duplicate rows are removed to avoid repeating information during analysis.

```python
df = df.drop_duplicates()
```

### Standardize categorical coumns

```python
for col in df.columns:
    unique_vals = df[col].astype(str).str.lower().unique()
    if set (unique_vals).issubset({'true', 'false'}):
        df[col] = df[col].astype(str).str.lower().map({'true':True, 'false': False})
```

### Convert boolean-like values

```python
for col in df.columns:
    unique_vals = df[col].astype(str).str.lower().unique()
    if set(unique_vals).issubset({'true', 'false'}):
        df[col] = df[col].astype(str).str.lower().map({'true':True, 'false': False})
```
